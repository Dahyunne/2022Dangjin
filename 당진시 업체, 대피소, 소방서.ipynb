```python
import pandas as pd
import numpy as np
import openpyxl
```

```python
sh_list = pd.read_excel('C://Users//Admin//Desktop//dang//당진시 대피소정보 위도경도.xlsx')
com_list = pd.read_excel('C://Users//Admin//Desktop//dang//업체 위도경도.xlsx')
fire_list = pd.read_excel('C://Users//Admin//Desktop//dang//당진시 소방서정보 위도경도.xlsx')
```

```python
import folium
map = folium.Map(location=[36.925, 126.7], zoom_start=11)
map
```

```python
for loc in sh_list.index:
    tooltip = sh_list['대피소명'][loc],"수용인원: ",sh_list['수용인원'][loc],"명"
    folium.Marker([sh_list['lat'][loc],sh_list['lng'][loc]],
            tooltip=tooltip,
            icon=folium.Icon(color = "green")
    ).add_to(map)
```

```python
for loc in com_list.index:
    tooltip = com_list['업체명'][loc],"취급물질: ",com_list['취급물질'][loc], '사고대비물질 취급여부 : ', '', '주민대피물질 취급여부', ''
    folium.Marker(
        [com_list['lat'][loc],com_list['lng'][loc]],
        tooltip=tooltip,
        icon=folium.Icon(color = "blue")
    ).add_to(map)
    folium.Circle(
        location=[com_list['lat'][loc],com_list['lng'][loc]],
        radius=500,
        color="#eb9e34",
        fill_color="red",
    ).add_to(map)
```

```python
for loc in fire_list.index:
    tooltip = fire_list['이름'][loc]
    folium.Marker(
        [fire_list['lat'][loc],fire_list['lng'][loc]],
        tooltip=tooltip,
        icon=folium.Icon(color = "red")
    ).add_to(map)
```
